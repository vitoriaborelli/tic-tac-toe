# Tic Tac Toe Game ##

```python
# BOARD DISPLAY
def display(board):
    print("Current game:\n")
    print(f' {board[0]} | {board[1]} | {board[2]}')
    print('-----------')
    print(f' {board[3]} | {board[4]} | {board[5]}')
    print('-----------')
    print(f' {board[6]} | {board[7]} | {board[8]}')
```

```python
# READING PLAYERS SYMBOL CHOICE
def initializing_game():
    
    player1 = ''
    player2 = ''
    
    print("Welcome to Vitoria's Tic Tac Toe!!")
    while player1 != 'X' and player1 != 'O':
        player1 = input("Player 1, would you like to be X or O? ").upper()
        if player1 == 'X':
            player2 = 'O'
            print("Player 1 is X and player 2 is O. Let's begin! Player 1 goes first!")
        elif player1 == 'O':
            player2 = 'X'
            print("Player 1 is O and player 2 is X. Let's begin! Player 1 goes first!")
        else:
            print("That's not a valid option, please choose again.")
            
    return (player1, player2)
```

```python
# READING LINE AND COLUMN OF CHOICE TO PLACE THE SYMBOL
def input_choice(board, player):
    
    line_choice = 0
    column_choice = 0
    position = 10 # ARBITRARY VALUE
    
    # DEFINING WHOSE PLAYER IS THE TURN
    if player%2 == 0:          
        turn = 'Player 2'
    else:
        turn = 'Player 1'
        
    # CHOOSING LINE
    while line_choice not in ['1', '2', '3']:
        line_choice = input(f"{turn}, at which line would you like to place your symbol? (1-3) ")
        if line_choice not in ['1', '2', '3']:
            print("This line does not exist, please choose again between given options.")
        
    # CHOOSING COLUMN
    while column_choice not in ['1', '2', '3']:
        column_choice = input(f"{turn}, at which column would you like to place your symbol? (1-3) ")
        if column_choice not in ['1', '2', '3']:
            print("This column does not exist, please choose again between given options.")
        
    # FINDING WHAT IS THE POSITION ON THE LIST
    if (line_choice, column_choice) == ('1','1'):
        position = 0
    elif (line_choice, column_choice) == ('1','2'):
        position = 1
    elif (line_choice, column_choice) == ('1','3'):
        position = 2
    elif (line_choice, column_choice) == ('2','1'):
        position = 3
    elif (line_choice, column_choice) == ('2','2'):
        position = 4
    elif (line_choice, column_choice) == ('2','3'):
        position = 5
    elif (line_choice, column_choice) == ('3','1'):
        position = 6
    elif (line_choice, column_choice) == ('3','2'):
        position = 7
    elif (line_choice, column_choice) == ('3','3'):
        position = 8        
            
    return position
    
```

```python
# CHECKING IF POSITION IS OCCUPIED AND, IN NEGATIVE CASE, PLACING SYMBOL
def occupy_position(board, position, player,player1, player2):
    if board[position] != ' ':
        print("This position is not available, choose one that is free.")
    else:
        # DEFINING WHAT IS THE SYMBOL BASED ON PLAYER ON TURN
        if player%2 == 0:          
            board[position] = player2
        else:
            board[position] = player1
        player += 1
        
    return (board, player)
```

```python
# CHECKING AFTER EVERY MOVE IF THE GAME IS STILL ON, TIED OR SOMEONE WON
def check_result(board, player1, player2):
    
    winner = ''
    # VICTORY POSSIBILITIES
    if board[0] == board[1] == board[2] != ' ':
        end = True
        winner = board[0]
    elif board[3] == board[4] == board[5] != ' ':
        end = True
        winner = board[3]
    elif board[6] == board[7] == board[8] != ' ':
        end = True
        winner = board[6]
    elif board[0] == board[3] == board[6] != ' ': 
        end = True
        winner = board[0]
    elif board[1] == board[4] == board[7] != ' ': 
        end = True
        winner = board[1]
    elif board[2] == board[5] == board[8] != ' ':
        end = True
        winner = board[2]
    elif board[0] == board[4] == board[8] != ' ':
        end = True
        winner = board[0]
    elif board[3] == board[4] == board[6] != ' ':
        end = True
        winner = board[3]
        
    # TIE (ALL CELLS FULL BUT NO WINNER)
    elif ' ' not in board:
        end = True
        display(board)
        print("\nIt's a tie! Nobody won.")
        
    # GAME ON
    else:
        end = False
        
    # DEFINING WINNER
    if winner == player1:
        display(board)
        print("\nCongratulations player 1! You're the winner of this round!")
    elif winner == player2:
        display(board)
        print("\nCongratulations player 2 You're the winner of this round!")
        
    return end
```

```python
# CHECKING IF USER WANTS TO PLAY AGAIN
def play_again():
    while True:
        again = input("Would you like to play one more time? (Y/N) ").upper()
        if again == 'Y':
            return True
        elif again == 'N':
            return False
        else:
            print("That is not one of the options.")
            continue
```

```python
from IPython.display import clear_output

# FINAL FUNCTION
def tic_tac_toe():
    
    # INITIALIZING BOARD
    board = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    # INITIALIZING GAME
    (player1, player2) = initializing_game()
    clear_output()
    
    end = False
    again = True
    player = 1
    
    while end == False and again == True:
        display(board)
        position = input_choice(board, player)
        clear_output()
        (board, player) = occupy_position(board, position, player, player1, player2)
        end = check_result(board, player1, player2)
        
        if end == True:
            again = play_again()
            player = 1
            clear_output()
    
```

```python
tic_tac_toe()
```
